## Check Outlier

```python
# List to store the non-categorical column names
non_categorical_columns = []

# Iterate over each column in the DataFrame
for column in df.columns:
    # Check if the column is non-categorical (numeric)
    if df[column].dtype != 'object':
        non_categorical_columns.append(column)

# Create the box plots
for column in non_categorical_columns:
    plt.figure()
    sns.boxplot(x='Sex', y=column, data=df)
    plt.title(f'Box Plot: {column}')
    plt.show()
```

## Check Data Distributed in different columns

```python
# Count of unique values for categorical columns
categorical_counts = df.select_dtypes(include='object').nunique()

# Histograms for numerical columns
numerical_cols = df.select_dtypes(include='number').columns
df[numerical_cols].hist(bins=20, figsize=(12, 8))
plt.suptitle('Histograms of Numerical Columns', y=1.02)
plt.show()

# Bar chart for a categorical column
categorical_column = 'Sex'
categorical_counts = df[categorical_column].value_counts()
plt.bar(categorical_counts.index, categorical_counts.values)
plt.title(f'Distribution of {categorical_column}')
plt.xlabel(categorical_column)
plt.ylabel('Count')
plt.show()
```

## Check Categorical Columns

```python
# Check the data types of each column in the DataFrame
data_types = df.dtypes

# Filter columns that have 'object' data type, which indicates categorical variables
categorical_columns = data_types[data_types == 'object'].index.tolist()

# Display the list of categorical columns
print("Categorical Columns:")
print(categorical_columns)
```

## Distribution of Categorical Columns

```python
# Filter columns that have 'object' data type, which indicates categorical variables
categorical_columns = df.select_dtypes(include='object').columns

# Calculate the frequency of each category in each categorical column
categorical_distribution = {}
for column in categorical_columns:
    categorical_distribution[column] = df[column].value_counts()

# Display the distribution of each categorical variable
for column, distribution in categorical_distribution.items():
    print(f"\nDistribution of {column}:")
    print(distribution)

    # Create a bar chart to visualize the distribution
    plt.figure(figsize=(8, 5))
    distribution.plot(kind='bar')
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()
```

## Survivel Rate Based on passenger class

```python
# Calculate the survival rate based on passenger class
survival_rate_by_class = df.groupby('Pclass')['Survived'].mean()

# Display the results
print("Survival Rate Based on Passenger Class:")
print(survival_rate_by_class)
```

## average fare paid by passengers who survived and those who did not

```python
# Calculate the average fare paid by passengers who survived
average_fare_survived = df[df['Survived'] == 1]['Fare'].mean()

# Calculate the average fare paid by passengers who did not survive
average_fare_not_survived = df[df['Survived'] == 0]['Fare'].mean()

# Display the results
print("Average Fare Paid by Passengers who Survived:", average_fare_survived)
print("Average Fare Paid by Passengers who Did Not Survive:", average_fare_not_survived)
```

## survival rate based on whether the passenger had a cabin or not

```python
# Fill missing cabin values with 'No Cabin'
df['Cabin'].fillna('No Cabin', inplace=True)

# Extract whether the passenger had a cabin or not
df['Cabin_Indicator'] = df['Cabin'].apply(lambda x: 'No Cabin' if x == 'No Cabin' else 'Cabin')

# Calculate the survival rate based on whether the passenger had a cabin or not
survival_rate_by_cabin = df.groupby('Cabin_Indicator')['Survived'].mean()

# Display the results
print("Survival Rate Based on Whether the Passenger had a Cabin:")
print(survival_rate_by_cabin)
```
